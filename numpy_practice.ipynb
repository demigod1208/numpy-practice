```python
import numpy as np
import random as r
import sys

list1 = [r.randint(1, 100) for i in range(10)]

print(list1)

print(f"Size of each element of list in bytes: {sys.getsizeof(list1[0])}")
print(f"Size of the whole list in bytes: {sys.getsizeof(list1)}")

array1d = np.array(list1,ndmin=1)
print(array1d)

print(f"Size of each element of the 1-D array in bytes: {array1d.itemsize}")
print(f"Size of the whole 1-D array in bytes: {array1d.nbytes}")

array2d = np.array(list1,ndmin=2)
print(array2d)

print(f"Size of each element of the 2-D array in bytes: {array2d.itemsize}")
print(f"Size of the whole 2-D array in bytes: {array2d.nbytes}")

array3d = np.array(list1,ndmin=3)
print(array3d)

print(f"Size of each element of the 3-D array in bytes: {array3d.itemsize}")
print(f"Size of the whole 3-D array in bytes: {array3d.nbytes}")
```
